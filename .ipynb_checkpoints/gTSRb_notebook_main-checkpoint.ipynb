## Step 0: Load the images

```python
# FUNCTION FOR LOADING IMAGES         
# trainImages, trainLabels = readTrafficSigns('GTSRB/Training')
# print len(trainLabels), len(trainImages)
# plt.imshow(trainImages[42])
# plt.show()
#
# have fun, Christian

import matplotlib.pyplot as plt
import csv

# function for reading the images
# arguments: path to the traffic sign data, for example './GTSRB/Training'
# returns: list of images, list of corresponding labels 
def readTrafficSigns(rootpath):
    '''Reads traffic sign data for German Traffic Sign Recognition Benchmark.

    Arguments: path to the traffic sign data, for example './GTSRB/Training'
    Returns:   list of images, list of corresponding labels'''
    images = [] # images
    labels = [] # corresponding labels
    dims = [] # images dims
    rois = [] # strict RoIs of signs
    # loop over all 42 classes
    for c in range(0,43):
        prefix = rootpath + '/' + format(c, '05d') + '/' # subdirectory for class
        gtFile = open(prefix + 'GT-'+ format(c, '05d') + '.csv') # annotations file
        gtReader = csv.reader(gtFile, delimiter=';') # csv parser for annotations file
        next(gtReader) # skip header
        # loop over all images in current annotations file
        for row in gtReader:
            images.append(plt.imread(prefix + row[0])) # the 1th column is the filename
            labels.append(row[7]) # the 8th column is the label
            dims.append( (row[1],row[2]) )
            rois.append( (row[3],row[4],row[5],row[6]) )
        gtFile.close()
    return images, labels, dims, rois
```

### Load training and testing images

```python
# Load training images
img, label, dims, rois = readTrafficSigns('./Training')

# Load testing images
img_tst, label_tst, dims_tst, rois_tst = readTrafficSigns('./Testing')
```

### Resize uniformly training and testing images, shuffle training set, and plot some examples

```python
import numpy as np
from sklearn.utils import shuffle
import skimage
#from skimage.util import crop
from skimage.transform import resize
from skimage.util import img_as_float32

img_size = 32

# Resize UNIFORMLY all training images
arr_train_img = np.asarray(img)
y_train = np.asarray(label)
X_train_img = arr_train_img
for i in range(0,arr_train_img.size):
  X_train_img[i] = resize(arr_train_img[i], (img_size,img_size) )

# Shuffle the training set (originally ordered by class after loading)
X_train_img, y_train = shuffle(X_train_img, y_train)

# Reformat training set as bi-dimensional array
n_train = X_train_img.shape[0]
input_dim = img_size*img_size*3
X_train = np.zeros( (n_train,input_dim) )
for i in range(0,X_train_img.size):
  X_train[i] = np.ravel( img_as_float32(X_train_img[i], True) )
print("Number of TRAINING examples = ",X_train.shape[0])

# Resize UNIFORMLY all testing images
arr_test_img = np.asarray(img_tst)
y_test = np.asarray(label_tst)
X_test_img = arr_test_img
for i in range(0,arr_test_img.size):
  X_test_img[i] = resize(arr_test_img[i], (img_size,img_size) )
    
# Reformat testing set as bi-dimensional array
n_test = X_test_img.shape[0]
input_dim = img_size*img_size*3
X_test = np.zeros( (n_test,input_dim) )
for i in range(0,X_test_img.size):
  X_test[i] = np.ravel( img_as_float32(X_test_img[i], True) )
print("Number of TESTING examples = ",X_test.shape[0])

# Visualize some training examples
class_names = label
for i in range(0,5):
  plt.figure()
  plt.imshow(X_train_img[i])
  plt.show()
  print(y_train[i])
```

```python
#Python script to display some images
def display_images_and_labels(images, labels):
    """Display the first image of each label."""
    unique_labels = set(labels)
    plt.figure(figsize=(15, 15))
    i = 1
    for label in unique_labels:
        # Pick the first image for each label.
        image = images[labels.index(label)]
        plt.subplot(8, 8, i)  # A grid of 8 rows x 8 columns
        plt.axis('off')
        plt.title("Label {0} ({1})".format(label, labels.count(label)))
        i += 1
        _ = plt.imshow(image)
    plt.show()
    
display_images_and_labels(img, label)
```

### Data distribution

```python
# Load sign names file
sign_names = pd.read_csv("signnames.csv")
sign_names.set_index("ClassId")

sign_names.head(n=3)
```

```python
# Group images in training set
def group_img_id_to_lbl(lbs_ids, lbs_names):    
    """ Utility function to group images by label """
    arr_map = []
    for i in range(0, lbs_ids.shape[0]):
        label_id = lbs_ids[i]
        label_name = lbs_names[lbs_names["ClassId"] == label_id]["SignName"].values[0]
        arr_map.append({"img_id": i, "label_id": label_id, "label_name": label_name})
    return pd.DataFrame(arr_map)

X_train_show = X_train.astype('int64')
X_test_show = X_test.astype('int64')
y_train_show = y_train.astype('int64')
y_test_show = y_test.astype('int64')
X_train_id_to_label = group_img_id_to_lbl(y_train_show, sign_names)

# print those mappings
X_train_id_to_label
```

```python
def group_img_id_to_lb_count(img_id_to_lb):
    """ Returns a pivot table table indexed by label id and label name, where the aggregate function is count """
    return pd.pivot_table(img_id_to_lb, index=["label_id", "label_name"], values=["img_id"], aggfunc='count')

# Group by label id to understand the distribution
X_train_group_by_label_count = group_img_id_to_lb_count(X_train_id_to_label)

# Count the number of images for each class and plot the results
X_train_group_by_label_count.plot(kind='bar', figsize=(15, 7))

X_train_group_by_label_count.head(n=5)
```

### Visualization of trainng set image

```python
# Plot 5 images from each of these classes
def show_image_list(img_list, img_labels, title, cols=2, fig_size=(15, 15), show_ticks=True):
    """ Utility function to show us a list of traffic sign images """
    img_count = len(img_list)
    rows = img_count // cols
    cmap = None
    fig, axes = plt.subplots(rows, cols, figsize=fig_size)
    for i in range(0, img_count):
        img_name = img_labels[i]     
        img = img_list[i]
        if len(img.shape) < 3 or img.shape[-1] < 3:
            cmap = "gray"
            img = np.reshape(img, (32, 32, 3))
        if not show_ticks:            
            axes[i].axis("off")
        axes[i].imshow(img, cmap=cmap)

    fig.suptitle(title, fontsize=12, fontweight='bold', y = 0.6)
    fig.tight_layout()
    plt.show()
    return

def show_random_dataset_images(group_label, imgs, to_show=5):
    """ This function takes a DataFrame of items group by labels as well as a set of images and randomly selects to_show images to display """
    for (lid, lbl), group in group_label:
        # print("[{0}] : {1}".format(lid, lbl))
        rand_idx = np.random.randint(0, high=group['img_id'].size, size=to_show, dtype='int')
        selected_rows = group.iloc[rand_idx]

        selected_img = list(map(lambda img_id: imgs[img_id], selected_rows['img_id']))
        selected_labels = list(map(lambda label_id: label_id, selected_rows['label_id']))
        show_image_list(selected_img, selected_labels, "{0}: {1}".format(lid, lbl), cols=to_show, fig_size=(7, 7), show_ticks=False)

X_train_group_by_label = X_train_id_to_label.groupby(["label_id", "label_name"])
img_per_class = 5

show_random_dataset_images(X_train_group_by_label, X_train_show)  
```

## Step 2: Comparision of algorithms: MLP, Random Forest, ConvNet

### MLP neural-network
Initially, we quick check of MLP scores and loss with the given default settings.
#### 1. The 1st MLP test with default settings

```python
# Data Preparation for MLP
X_train_mlp = X_train.reshape(26640,3072).astype('float32')
X_test_mlp = X_test.reshape(12569,3072).astype('float32')
y_train_mlp = y_train.reshape(26640,).astype('float32')
y_test_mlp = y_test.reshape(12569,).astype('float32')

print("training_shape for MLP: ", X_train_mlp.shape,y_train_mlp.shape)
print("testing_shape for MLP: ", X_test_mlp.shape,y_test_mlp.shape)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=30, alpha=1e-4,
                    solver='sgd', verbose=10, tol=1e-4, random_state=1,
                    learning_rate_init=0.03)

# Train the MLP classifier on training dataset
mlp.fit(X_train_mlp, y_train_mlp)

# Plot the LEARNING CURVE
plt.title("Evolution of TRAINING ERROR during training")
plt.xlabel("Iterations (epochs)")
plt.ylabel("TRAINING ERROR")
plt.plot(mlp.loss_curve_)
plt.show()

# Evaluate acuracy on test and training data
print("Training set score: %f" % mlp.score(X_train_mlp, y_train_mlp))
print("Test set score: %f" % mlp.score(X_test_mlp, y_test_mlp))
y_pred_mlp = mlp.predict(X_test_mlp)
print("Detailed classification report:")
print(classification_report(y_test_mlp, y_pred_mlp))

# Display CONFUSION MATRIX on TEST set
print("CONFUSION MATRIX below")
confusion_matrix(y_test_mlp, y_pred_mlp)
```

Assume that the first given MLP test does not perform perfectly, we increase the number of iterations with additional early_stopping parameter.

We then use grid-search and cross-validation method to find an optimized combination of learning hyperpaprameters and to avoid the over-fitting. Due to the long training time of the MLP model, we will only show the best set of results here.

```python
%%time
# Define the parameters and values to be tested
param_grid = [
    {'hidden_layer_sizes': [(10,), (50,), (100,)], 
    'learning_rate_init':[0.003, 0.03, 0.1],
    'alpha': [0.0001, 0.001]}
    ]

# Cross-validation grid-search (for finding best possible accuracy)
mlp = GridSearchCV( MLPClassifier(max_iter=100,
                    solver='sgd', tol=1e-4, random_state=1,
                    learning_rate_init=0.03, early_stopping=True, validation_fraction=0.4),
                    param_grid, cv=3, scoring='accuracy')

mlp.fit(X_train_mlp, y_train_mlp)
print("Best parameters set found on development set:")
print()
print(mlp.best_params_)
print()
print("Grid scores on development set:")
print()
means = mlp.cv_results_['mean_test_score']
stds = mlp.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, mlp.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
           % (mean, std * 2, params))
```

### 2. The 2nd MLP test with best_param

```python
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=100, alpha=1e-3,
                    solver='sgd', tol=1e-4, random_state=1,
                    learning_rate_init=0.03)

# Train the MLP classifier on training dataset
mlp.fit(X_train_mlp, y_train_mlp)

# Plot the LEARNING CURVE
plt.title("Evolution of TRAINING ERROR during training")
plt.xlabel("Iterations (epochs)")
plt.ylabel("TRAINING ERROR")
plt.plot(mlp.loss_curve_)
plt.show()

# Evaluate acuracy on test and training data
print("Training set score: %f" % mlp.score(X_train_mlp, y_train_mlp))
print("Test set score: %f" % mlp.score(X_test_mlp, y_test_mlp))
y_pred_mlp = mlp.predict(X_test_mlp)
print("Detailed classification report:")
print(classification_report(y_test_mlp, y_pred_mlp))

# Display CONFUSION MATRIX on TEST set
print("CONFUSION MATRIX below")
confusion_matrix(y_test_mlp, y_pred_mlp)
```

### CNN

CNN as an efficient way for image classification will be applied here. 

```python
# Data Preparation for ConvNet
X_train_cnn = X_train.reshape(26640,32,32,3).astype('float32')
X_test_cnn = X_test.reshape(12569,32,32,3).astype('float32')
y_train_cnn = y_train.reshape(26640,1).astype('float32')
y_test_cnn = y_test.reshape(12569,1).astype('float32')

print("training_shape for ConvNet: ", X_train_cnn.shape,y_train_cnn.shape)
print("testing_shape for ConvNet: ", X_test_cnn.shape,y_test_cnn.shape)
```

#### 1. The 1st CNN test with default settings
Firstly, without overfitting prevention methods like dropout and regularization, the performance of ConvNet model is tested by the accuracies and losses curves. 

```python
# Network params
batch_size = 32    # in each iteration, we consider 32 training examples at once
num_epochs = 20    # we iterate 20 times over the entire training set
kernel_size = 5    # we will use 5x5 kernels throughout
pool_size = 2      # we will use 2x2 pooling throughout
conv_depth_1 = 6   # we will initially have 6 kernels in first conv. layer...
conv_depth_2 = 16  # ...switching to 16 after the first pooling layer
drop_prob_1 = 0.   # dropout after pooling with probability 0.
drop_prob_2 = 0.   # dropout in the FC layer with probability 0.
hidden_size = 128  # the FC layer will have 128neurons
weight_penalty = 0. # Factor for weights penalty
num_classes = 43    # Number of classes

model = Sequential()
# Conv [8] -> Pool (without dropout on the pooling layer)
model.add( Conv2D(conv_depth_1, (kernel_size,kernel_size), padding='same', activation='relu', data_format="channels_last", input_shape=( 32, 32, 3)) )
model.add( MaxPooling2D(pool_size=(pool_size, pool_size)) )
model.add( Dropout(drop_prob_1) )

# Conv [16] -> Pool (without dropout on the pooling layer)
model.add( Conv2D(conv_depth_2, (kernel_size,kernel_size), padding='same', activation='relu') )
model.add( MaxPooling2D(pool_size=(pool_size, pool_size)) )
model.add( Dropout(drop_prob_1) )

# Now flatten to 1D, apply FC -> ReLU (without dropout) -> softmax
model.add( Flatten() )
model.add( Dense(hidden_size, activation='relu', kernel_regularizer=regularizers.l2(weight_penalty)) )
model.add( Dropout(drop_prob_2) )
model.add( Dense(num_classes, activation='softmax') )

print(model.summary())

# Compile Model
model.compile(loss='sparse_categorical_crossentropy', 
            optimizer='adam', 
            metrics=['accuracy'])

# Train Model
epochs = 8
history = model.fit(X_train_cnn, y_train_cnn, 
                    epochs=epochs, batch_size=64,
                    verbose=1, validation_split=0.4)

# Evaluate Model
for loss_name, loss_value in list(zip(model.metrics_names, model.evaluate(X_train_cnn, y_train_cnn, verbose=1))):
    print('The final {} on the TRAINING set is: {:.2f}.'.format(loss_name, loss_value)) 
for loss_name, loss_value in list(zip(model.metrics_names, model.evaluate(X_test_cnn, y_test_cnn, verbose=1))):
    print('The final {} on the TEST set is: {:.2f}.'.format(loss_name, loss_value)) # Evaluate the trained model on the test set!
```

```python
# Accuracy & Loss functions evolution
plt.figure(0)
plt.plot(history.history['accuracy'], label="Training accuracy")
plt.plot(history.history['val_accuracy'], label="Val accuracy")
plt.title("Accuracy Graph")
plt.xlabel("epochs")
plt.ylabel("accuracy (0,1)")
plt.legend()

plt.figure(1)
plt.plot(history.history['loss'], label="Training loss")
plt.plot(history.history['val_loss'], label="Val loss")
plt.title("Loss Graph")
plt.xlabel("epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

In the initial test of CNN, the accuracy of test set is much lower and the over-fitting may begin to occur at the 3rd epoch. In order to mitigate over-fitting, the dropout and regularization should be performed. After several trials of CNN architectures, the optimal CNN design with the greatest accuracy is shown as follows.

#### 2. The 2nd CNN test with optimal settings

```python
# The optimal CNN architecture
model = tf.keras.Sequential()
# Conv -> Pool (with dropout on the pooling layer)
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu', input_shape=X_train_cnn.shape[1:]))
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))

# Conv -> Pool (with dropout on the pooling layer)
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))

# Now flatten to 1D, apply FC -> ReLU (with dropout) -> softmax
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(43, activation='softmax'))

# Model display
model.summary()

# Compile Model
model.compile(loss='sparse_categorical_crossentropy', 
            optimizer='adam', 
            metrics=['accuracy'])

# Train Model
epochs = 8
history = model.fit(X_train_cnn, y_train_cnn, 
                    epochs=epochs, batch_size=64,
                    verbose=1, validation_split=0.4)

# Evaluate Model
for loss_name, loss_value in list(zip(model.metrics_names, model.evaluate(X_train_cnn, y_train_cnn, verbose=1))):
    print('The final {} on the TRAINING set is: {:.2f}.'.format(loss_name, loss_value)) 
for loss_name, loss_value in list(zip(model.metrics_names, model.evaluate(X_test_cnn, y_test_cnn, verbose=1))):
    print('The final {} on the TEST set is: {:.2f}.'.format(loss_name, loss_value)) # Evaluate the trained model on the test set!
```

```python
# Accuracy & Loss functions evolution
plt.figure(0)
plt.plot(history.history['accuracy'], label="Training accuracy")
plt.plot(history.history['val_accuracy'], label="Val accuracy")
plt.title("Accuracy Graph")
plt.xlabel("epochs")
plt.ylabel("accuracy (0,1)")
plt.legend()

plt.figure(1)
plt.plot(history.history['loss'], label="Training loss")
plt.plot(history.history['val_loss'], label="Val loss")
plt.title("Loss Graph")
plt.xlabel("epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

From the training results, we witness that ConvNet model can achieve the best performance topping the 98% test accuracy. Besides, we can see on the graph of Loss vs Epochs that the losses of training and validation set get closer as the epochs go by, which would indicate that overfitting is mitigated by dropout method and ConvNet model generalises well to extract features.
